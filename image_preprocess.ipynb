```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "Inference_results/Originals/Yaqian_original_image.jpg"
# img_path = "Inference_results/Originals/Yae_Miko_Avatar.png"
image = cv2.imread(img_path)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not open or find the image.")
else:
    # Display the image in a window
# Convert the image from BGR (OpenCV default) to RGB (Matplotlib uses RGB)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Optional: Hides the axes
    plt.show()
```
